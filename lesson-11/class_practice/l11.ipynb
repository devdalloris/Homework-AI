```python
import numpy as np

class DummyClassifier:
    def fit(self, X, y):
        return self
    
    def predict(self, X):
        N = X.shape[0]
        return np.zeros(N)
    
    def score(self, X, y):
        y_pred = self.predict(X)
        return (y_pred == y).sum() / y.shape[0]
    

dummy_clf = DummyClassifier()
dummy_clf.fit(X_train, y_train_5)

y_pred = dummy_clf.predict(X_train)
```
