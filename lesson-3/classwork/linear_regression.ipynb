### Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'G:\AI\AI\Homework AI\lesson-3\data\student_scores.csv')
```

```python
df.head()
```

```python
def model(Hours):
    return 9.5*Hours+5
y_pred = model(df['Hours'])
plt.scatter(df['Hours'], df['Scores'])
plt.plot(df['Hours'], y_pred)
plt.scatter(df['Hours'], y_pred, c='r')
plt.xlabel('Hours Studied')
plt.ylabel('Scores Obtained')
plt.show()
```

```python
y = df['Scores'].values
y_pred = model(df['Hours']).values
np.sqrt(np.sum((y-y_pred)**2))/len(y)
```

```python
import statistics 
var_x = statistics.variance(df['Hours'])
```

```python
cov_xy = statistics.covariance(df['Hours'], df['Scores'])
```

```python
statistics.correlation(df['Hours'], df['Scores'])
```

```python
beta_1 = cov_xy/var_x
beta_1
```

### Linear Regression with Scikit learn

```python
from sklearn.linear_model import LinearRegression
model =  LinearRegression()

X=df['Hours'].values
y=df['Scores'].values
X = X.reshape(-1,1)
# y = y.reshape(-1,1)
```

```python
model.fit(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict(X)
```
