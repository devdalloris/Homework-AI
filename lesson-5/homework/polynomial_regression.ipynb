```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, y, 'b.')
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([-3, 3, 0, 10])
plt.grid()
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(data=np.c_[X,y], columns=['x', 'y'])
df.to_excel('sample.xlsx', index=False)
```

```python
from sklearn.linear_model import LinearRegression
X2=X**2
X_all = np.c_[X,X2]

lin_reg = LinearRegression()
lin_reg.fit(X_all, y)
```

```python
y_pred = lin_reg.predict(X_all)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, y, 'b.')
plt.scatter(X, y_pred, c='green')
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([-3, 3, 0, 10])
plt.grid()
plt.show()
```
