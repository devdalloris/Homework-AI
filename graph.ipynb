```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # Used for creating dummy data if CSV load fails

# --- 1. Load the dataset ---
file_name = 'ctn_d.csv'
try:
    df = pd.read_csv(file_name)
    print(f"Dataset '{file_name}' loaded successfully. Shape: {df.shape}")
    print("\nFirst 5 rows of the dataset:")
    print(df.head())
    print("\nDataFrame Info:")
    df.info()
except FileNotFoundError:
    print(f"Error: '{file_name}' not found. Please ensure the file is in the correct directory.")
    print("Creating a hypothetical DataFrame for demonstration purposes to show pie chart functionality.")
    # Create a dummy DataFrame if the actual file is not found
    np.random.seed(42)
    data = {
        'number': np.random.randint(5, 20, size=100), # Total items (e.g., plants detected in an area)
        'num_weeds': np.random.randint(0, 10, size=100) # Number of weeds within those items
    }
    # Ensure num_weeds is not greater than number for dummy data
    df = pd.DataFrame(data)
    df['num_weeds'] = df.apply(lambda row: min(row['num_weeds'], row['number']), axis=1)
    print("\nUsing a hypothetical DataFrame:")
    print(df.head())
    print("\nDataFrame Info (Hypothetical):")
    df.info()

# --- 2. Prepare data for the Pie Chart ---
# Assuming 'number' represents the total count of items (e.g., total plants in an image)
# and 'num_weeds' represents the count of weeds within those items.
# We want to show the proportion of 'Total Weeds' vs 'Total Non-Weeds'.

# Ensure columns are numeric; coerce errors will turn non-numeric values into NaN
df['number'] = pd.to_numeric(df['number'], errors='coerce')
df['num_weeds'] = pd.to_numeric(df['num_weeds'], errors='coerce')

# Drop rows where critical data for calculation might be missing
df.dropna(subset=['number', 'num_weeds'], inplace=True)

# Calculate total 'weeds' and 'non-weeds' across the dataset
total_weeds = df['num_weeds'].sum()
total_non_weeds = (df['number'] - df['num_weeds']).sum()

# Handle cases where 'total_non_weeds' might be negative due to data inconsistencies
if total_non_weeds < 0:
    print("\nWarning: 'total_non_weeds' calculation resulted in a negative value.")
    print("This might indicate that 'num_weeds' sometimes exceeds 'number', or 'number' is not a total count.")
    print("Adjusting 'total_non_weeds' to 0 for plotting, but please review your data's meaning.")
    total_non_weeds = 0 # Prevent negative values in pie chart
    
# If both sum to zero, the pie chart cannot be created
if total_weeds == 0 and total_non_weeds == 0:
    print("\nBoth 'Total Weeds' and 'Total Non-Weeds' sum to zero. Cannot create a pie chart.")
    exit()

# Combine into a Series for the pie chart.
# The pie chart will partition based on these two values.
data_for_pie = pd.Series({
    'Total Weeds': total_weeds,
    'Total Non-Weeds': total_non_weeds
})

# Filter out categories with zero values to avoid plotting empty slices
data_for_pie = data_for_pie[data_for_pie > 0]


# --- 3. Create the Pie Chart ---
plt.figure(figsize=(10, 8)) # Set figure size for good presentation

plt.pie(data_for_pie,
        labels=data_for_pie.index, # Labels for each slice ('Total Weeds', 'Total Non-Weeds')
        autopct='%1.1f%%',   # Format percentages to one decimal place. This directly describes the proportion of 'number' (total items) that are 'num_weeds' (total weeds) or 'non-weeds'.
        startangle=90,       # Start the first slice at the top (visually appealing)
        colors=['darkred', 'forestgreen'], # Thematic colors: Dark red for weeds (problematic), Forest green for non-weeds (healthy crops/soil)
        wedgeprops={'edgecolor': 'black'}) # Add a black border around each slice for clarity

plt.title('Proportion of Total Weeds vs. Total Non-Weeds (from ctn_d.csv)', fontsize=16, fontweight='bold')
plt.axis('equal') # Ensures the pie chart is drawn as a perfect circle.
plt.tight_layout() # Adjusts plot parameters for a tight layout.
plt.show() # Display the chart

# --- 4. Explanation of the Pie Chart ---
print("\n--- Interpretation of the Pie Chart ---")
print("This pie chart visually represents the overall proportion of 'Total Weeds' versus 'Total Non-Weeds' across your entire 'ctn_d.csv' dataset.")
print("The 'Total Weeds' slice (colored dark red) shows the aggregated sum of all values in the 'num_weeds' column. The 'Total Non-Weeds' slice (colored forest green) represents the aggregated sum of the difference between 'number' and 'num_weeds' for all entries. This provides insight into the count of desired elements (like cotton plants) relative to the weeds.")
print("The percentages displayed on the slices directly indicate the proportion of the total recorded items that fall into the 'weed' or 'non-weed' category. This chart offers a quick, thematic summary of the overall weed infestation level relative to the total number of items recorded in your dataset, using colors chosen to reflect the agricultural context.")
```
