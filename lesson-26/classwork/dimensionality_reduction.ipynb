### Principal Component Analysis

```python
from sklearn.datasets import fetch_openml
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784')
```

```python
X, y = mnist.data.values, mnist.target.values
```

```python
import numpy as np

target_names = np.unique(y)
```

```python
target_names
```

```python
forest_clf = RandomForestClassifier(random_state=42)
forest_clf.fit(X, y)
```

```python
mask = forest_clf.feature_importances_!=0
```

```python
X[:, mask].shape
```

```python
plt.imshow(X[0].reshape(28,28), cmap='binary')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_reduced = pca.fit_transform(X)
```

```python
pca.explained_variance_ratio_
```

```python
X.shape
```

```python
X_reduced.shape
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan']
for target, color, label in zip(target_names, colors, target_names.astype(str)):
    ax.scatter(X_reduced[y==target, 0], X_reduced[y==target, 1], X_reduced[y==target, 2], c=color, label=label)   
ax.legend()       
plt.show() 
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=3, random_state = 42)
X_reduced_tsne = tsne.fit_transform(X[:10000])
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan']
for target, color, label in zip(target_names, colors, target_names.astype(str)):
    ax.scatter(X_reduced_tsne[y[:10000]==target, 0], X_reduced_tsne[y[:10000]==target, 1], X_reduced_tsne[y[:10000]==target, 2], c=color, label=label)   
ax.legend()       
plt.show() 
```
